# Make soybean_weekly.csv

```python
# EMD installation

# Imports
import pandas as pd

# Pull soybean data
data_url = 'https://raw.githubusercontent.com/ptraver/data/main/bean_daily_V2.csv'
#data_url = '/content/soybean_weekly_1.csv'
soybean_df = pd.read_csv(data_url, parse_dates=True, dayfirst=True, index_col='date')

# fix bad name
soybean_df.columns = ['value']

# take only 1974 - 2017
soybean_df = soybean_df['1974-01-01':'2017-12-31']
```

```python
# downsample to weekly
soybean_df = soybean_df.value.resample('W').mean()
```

```python
# Save dataframe
soybean_df.to_csv('soybean_weekly.csv')
soybean_df.head()
```
